```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR= Q3-Q1
    lower_range= Q1-(1.5*IQR)
    upper_range= Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
lrsales,ursales=remove_outlier(sales['SALES'])
sales['SALES']=np.where(sales['SALES']>ursales,ursales,sales['SALES'])
sales['SALES']=np.where(sales['SALES']<lrsales,lrsales,sales['SALES'])

lrsales,ursales=remove_outlier(sales['QUANTITYORDERED'])
sales['QUANTITYORDERED']=np.where(sales['QUANTITYORDERED']>ursales,ursales,sales['QUANTITYORDERED'])
sales['QUANTITYORDERED']=np.where(sales['QUANTITYORDERED']<lrsales,lrsales,sales['QUANTITYORDERED'])

lrsales,ursales=remove_outlier(sales['PRICEEACH'])
sales['PRICEEACH']=np.where(sales['PRICEEACH']>ursales,ursales,sales['PRICEEACH'])
sales['PRICEEACH']=np.where(sales['PRICEEACH']<lrsales,lrsales,sales['PRICEEACH'])


lrsales,ursales=remove_outlier(sales['MSRP'])
sales['MSRP']=np.where(sales['MSRP']>ursales,ursales,sales['MSRP'])
sales['MSRP']=np.where(sales['MSRP']<lrsales,lrsales,sales['MSRP'])



```

```python
# defining categorical variables
categorical_vars = ['STATUS', 'PRODUCTLINE', 'DEALSIZE','COUNTRY']  # Add other categorical variables as needed

# Create a figure with subplots for each categorical variable
fig, axs = plt.subplots(len(categorical_vars), 1, figsize=(10, 6 * len(categorical_vars)))

# Loop through each categorical variable and create a bar plot
for i, var in enumerate(categorical_vars):
    # Count the frequency of each category
    category_counts = sales[var].value_counts()
    
    # Create a bar plot using Matplotlib Axes
    axs[i].bar(category_counts.index, category_counts.values, color='skyblue')
    axs[i].set_title(f'Distribution of {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frequency')
    axs[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Adjust layout

plt.show()
```
